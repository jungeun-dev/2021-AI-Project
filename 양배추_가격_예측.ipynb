```python
import torch
import pandas as pd

torch.manual_seed(1)

train = pd.read_csv('../input/2021-ai-w3-p1/train.csv')
test = pd.read_csv('../input/2021-ai-w3-p1/test.csv')
submit = pd.read_csv('../input/2021-ai-w3-p1/submit_sample.csv')

train
```

```python
import numpy as np

x_train = torch.FloatTensor(np.array(train)[: , 1 : -1])
y_train = torch.FloatTensor(np.array(train)[: , [-1]])

x_test = torch.FloatTensor(np.array(test)[: , 1: ])

x_train
y_train
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train = torch.FloatTensor(x_train)
x_test = torch.FloatTensor(x_test)
```

```python
# layer1 = torch.nn.Linear(4,16)
# layer2 = torch.nn.Linear(16,4)
# layer3 = torch.nn.Linear(4,1)
# sigmoid = torch.nn.Sigmoid()

layer = torch.nn.Linear(4,1)

model = torch.nn.Sequential(layer)
# model = torch.nn.Sequential(layer1, layer2, layer3)
```

```python
import torch.optim as optim

optimizer = optim.SGD(model.parameters(), lr = 0.01)
loss = torch.nn.MSELoss()

epoch = 60000
for e in range(epoch + 1):
    hypothesis = model(x_train)
    
    optimizer.zero_grad()
    cost = loss(hypothesis, y_train)
    

    cost.backward()
    optimizer.step()
    
    if e % 1000 == 0:
        print("epoch : ", e, "cost : ", cost.item() )
    
```

```python
predict = model(x_test)
print(predict)
```

```python
submit['Expected'] = predict.detach().numpy()
```

```python
submit.to_csv('submission.csv',index = False)

s = pd.read_csv('submission.csv')
s
```
