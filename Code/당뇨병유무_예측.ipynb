```python
import torch
import pandas as pd

torch.manual_seed(1)

train = pd.read_csv('../input/2021-ai-w4-p1/train.csv')
test = pd.read_csv('../input/2021-ai-w4-p1/test.csv')
submit = pd.read_csv('../input/2021-ai-w4-p1/submit.csv')

train
```

```python
x_train = train.drop('Diabetes', axis = 1)
y_train = train['Diabetes']

test
```

```python
#í‘œì¤€í™”
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(test)

x_train
```

```python
x_train = torch.FloatTensor(x_train)
y_train = torch.FloatTensor(y_train)
x_test = torch.FloatTensor(x_test)
```

```python
import torch.optim as optim
x_train.shape

feature = 8
lr = 0.01
epochs = 30000

W = torch.zeros((feature,1) , requires_grad = True)
b = torch.zeros(1, requires_grad = True)

loss = torch.nn.BCELoss()
optimizer = optim.SGD([W,b], lr = lr)
```

```python
for e in range(epochs + 1):
    hypothesis = torch.sigmoid(x_train.matmul(W) + b)
    
    cost = loss(hypothesis , y_train.unsqueeze(1))
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if e % 1000 == 0:
        print("epoch : ", e, "cost : ", cost.item() )
    
```

```python
submit.info()
```

```python
hypothesis = torch.sigmoid(x_test.matmul(W) + b)
predict = hypothesis >= torch.FloatTensor([0.5])
for i in range(len(predict)):
    submit['Diabetes'][i] = predict[i]
print(submit)
```
